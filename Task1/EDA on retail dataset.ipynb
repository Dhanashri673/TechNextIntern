# Exploratory Data Analysis

# Gender

```python
df.columns
```

```python
#Plotting the count distribution of customers by gender with labels on bars

plt.figure(figsize=(6,4))
ax = sns.countplot(x='Gender', data = df, palette="pastel")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Gender Distribution of Customers")
plt.show()
```

```python
# Plotting total sales amount by gender using a bar chart

sales_by_gender = df.groupby('Gender')['Amount'].sum().reset_index()
sns.barplot(x='Gender', y='Amount', data=sales_by_gender, palette="muted")
plt.title("Total Sales Amount by Gender")
plt.show()
```

This graph is same as above count plot based on Gender graph. but it uses ax.bar_label(bars) to add the count as a label inside each bar(like 7832 and 3407)

```python
# Grouping data by gender to calculate total sales amount and sorting in descending order  

df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

This tells us that Females spent more than Males based on total Amount.

```python
# Creating a bar plot of total sales amount by gender sorted in descending order

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y= 'Amount', data= sales_gen)
```

This graph tells that female spent more money than males

# Age

```python
df.columns
```

```python
# Plotting the distribution of customers across age groups with gender-wise comparison

plt.figure(figsize=(8,5))
sns.countplot(x='Age Group', hue='Gender', data=df, palette="Set2")
plt.title("Age Group Distribution with Gender")
plt.show()
```

Most of the customers are Females in the age group 26-35, followed by 18-25 and 36-45. Here shows that young to middle-aged females are the dominant buyers in the dataset. And the plot didn't follow the correct age sequence because the column is string based, and Seaborn used alphabetical order.

```python
# Creating a bar plot of total sales amount by different age groups

sales_by_age = df.groupby('Age Group')['Amount'].sum().reset_index()
sns.barplot(x='Age Group', y='Amount', data=sales_by_age, palette="coolwarm")
plt.title("Total Sales by Age Group")
plt.show()
```

from this graph, we can see that majority of customers are young adults i.e., 26-35 years then this dataset should focus on products appealing to 26-35, followed by 36-45 and 18-25 groups. Teenagers 0-17 and seniors 55+ are a very small portion of the customer base.

# State

```python
df.columns
```

```python
# Plotting the top 10 states with the highest total sales amount

top_states = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='State', y='Amount', data=top_states, palette="viridis")
plt.xticks(rotation=45)
plt.title("Top 10 States by Sales")
plt.show()
```

This bar chart shows the number of orders placed from different states in India.The top 3 states contributing the most orders are Uttar Pradesh, Maharashtra, and Karnataka.
States like Jharkhand, Bihar, and Uttarakhand are among the lowest in terms of order count.

# Marital Status

```python
# Creating a bar plot of total sales amount based on marital status

sales_by_marital = df.groupby('Marital_Status')['Amount'].sum().reset_index()
sns.barplot(x='Marital_Status', y='Amount', data=sales_by_marital, palette="Set1")
plt.title("Sales by Marital Status")
plt.show()
```

# occupation

```python
# Plot total sales amount by occupation using a bar chart

occupation_sales = df.groupby('Occupation')['Amount'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='Occupation', y='Amount', data=occupation_sales, palette="plasma")
plt.xticks(rotation=45)
plt.title("Sales by Occupation")
plt.show()
```

# Product Category

```python
# Plot top 10 product categories by total sales amount using a bar chart

top_products = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='Product_Category', y='Amount', data=top_products, palette="cubehelix")
plt.xticks(rotation=45)
plt.title("Top 10 Product Categories by Sales")
plt.show()
```

# Order vs Amount

```python
# Plot scatter plot to visualize relationship between number of orders and sales amount

plt.figure(figsize=(6,6))
sns.scatterplot(x='Orders', y='Amount', data=df, alpha=0.6, color="red")
plt.title("Orders vs Amount")
plt.show()
```

# Top 10 products by sales amount

```python
# Plot top 10 products by total sales amount using a bar chart

top_products = df.groupby('Product_ID')['Amount'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='Product_ID', y='Amount', data=top_products, palette="magma")
plt.xticks(rotation=45)
plt.title("Top 10 Products by Sales Amount")
plt.show()
```

This graph groups sales data by product, finds the 10 most-ordered products, and visualizes them in a bar chart.Shows which specific products are the core drivers of customer demand.

# Zone-wize sales

```python
# Plot total sales amount by zone using a bar chart

zone_sales = df.groupby('Zone')['Amount'].sum().reset_index()
sns.barplot(x='Zone', y='Amount', data=zone_sales, palette="Spectral")
plt.title("Sales by Zone")
plt.show()
```

```python
# Calculate total sales (revenue)
total_sales = df['Amount'].sum()
print("Total Sales (Revenue):", total_sales)
```

```python
# Calculate revenue per product category
revenue_per_category = df.groupby('Product_Category')['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False)

print(revenue_per_category)
```

```python
# Calculate Average Order Value (AOV)
total_sales = df['Amount'].sum()
total_orders = df['Orders'].sum()
aov = total_sales / total_orders

print("Average Order Value (AOV):", round(aov, 2))
```

```python
# Top 10 Products by Total Sales Amount
top_products = df.groupby('Product_ID')['Amount'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='Product_ID', y='Amount', data=top_products, palette="magma")
plt.xticks(rotation=45)
plt.title("Top 10 Products by Sales Amount")
plt.show()

# Top 10 Products by Number of Orders
top_orders = df.groupby('Product_ID')['Orders'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='Product_ID', y='Orders', data=top_orders, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Top 10 Products by Number of Orders")
plt.show()
```
